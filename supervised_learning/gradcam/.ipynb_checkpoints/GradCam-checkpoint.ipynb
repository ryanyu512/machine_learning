# Import Libraries

```python
#for data processing or visualization
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
import random
import math

#for deep learning module
import torch 
from torch import nn 
from torchsummary import summary
from torch.utils.data import DataLoader, Dataset 
from torchvision import datasets
from torchvision import transforms

#for tracking training/validation progress
from tqdm import tqdm 
```

# Configurations

```python
#define data directories
ROOT = '/Users/r/Desktop/git_repository/machine_learning/dataset/GradCAM-Dataset/'
CSV_DIR = ROOT + 'train.csv'

#define batch size for training
BATCH_SIZE = 16

#define learning rate for training
LR = 1e-4

#define number of epochs for training
EPOCH = 50

#define the device
if torch.backends.mps.is_available():
    DEVICE = 'mps'
elif torch.cuda.is_available():
    DEVICE = 'cuda'
else:
    DEVICE = 'cpu'

print(f'device: {DEVICE}')

#define class name
class_dict = {0: "cucumber", 1: "eggplant", 2: "mushroom"}
```

```python
#load data directories and corresponding labels
df = pd.read_csv(CSV_DIR)
#display the first 10 rows of data
df.head(10)
```

# Split Data into Training, Validation and Testing data

```python
def split_ids(data, train=.7, valid=0.1, test=0.2, seed = 0):
    """
    Args:
        image_id_list (int list): list of unique image ids
        train (float): train split size (between 0 - 1)
        valid (float): valid split size (between 0 - 1)
        test  (float): test split size (between 0 - 1)
    Outputs:
        train_set: list of training data
        valid_set: list of validation data
        test_set: list of testing data
    """
    
    list_copy = list(range(0, len(data)))
    random.Random(seed).shuffle(list_copy)
    
    #obtain the size of training, validation and testing data
    train_size = math.floor(len(list_copy) * train)
    valid_size = math.floor(len(list_copy) * valid)
    test_size = len(list_copy) - train_size - valid_size
    
    train_set = [None]*train_size
    if train + valid == 1.0:
        valid_size += test_size
        valid_set = [None]*valid_size
        test_set  = None
        test_size = 0
    else:
        valid_set = [None]*valid_size
        test_set  = [None]*test_size
    
    #split the data into training, validation and testing dataset
    idx = 0
    for cnt, i in enumerate(list_copy):
        
        if cnt == train_size or cnt == train_size + valid_size:
            idx = 0
            
        if cnt < train_size:
            train_set[idx]= (data['img_path'].iloc[i], 
                             data['label'].iloc[i])
        elif cnt >= train_size and cnt < train_size + valid_size:
            valid_set[idx]= (data['img_path'].iloc[i], 
                             data['label'].iloc[i])
        else:
            test_set[idx] = (data['img_path'].iloc[i], 
                             data['label'].iloc[i])
        idx += 1
        
    return train_set, valid_set, test_set
```

```python
train_data, valid_data, test_data = split_ids(df, train = 0.6, valid = 0.2, test = 0.2)

#check if the training and validation dataset are in correct proportion
N_data = len(df)
N_train = len(train_data)
N_valid = len(valid_data)
N_test = len(test_data)

print(f'Total number of data: {N_data}')
print(f'Total number of training data: {N_train} ({N_train/N_data :.2f})')
print(f'Total number of validation data: {N_valid} ({N_valid/N_data :.2f})')
print(f'Total number of validation data: {N_test} ({N_test/N_data :.2f})')
```

```python
#define data augmentation for training and validation

train_augs = transforms.Compose([ 
    transforms.ToTensor(),
    transforms.RandomRotation(degrees = (-10, 10)),
    transforms.RandomVerticalFlip(0.5),
    transforms.RandomHorizontalFlip(0.5),
    transforms.Normalize((0.485, 0.456, 0.406),   # using ImageNet norms
                         (0.229, 0.224, 0.225))
    ])

valid_augs = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.485, 0.456, 0.406), 
                         (0.229, 0.224, 0.225))
])
```

```python
#load dataset WITH data Augumentation

def load_dataset(data, augs):
    
    dataset = [None]*len(data)
    for i, (dir, label) in enumerate(data):
        img = plt.imread(ROOT + dir)
        img = augs(img)
        dataset[i] = (img, label)
        
    return dataset

#load training dataset
train_set = load_dataset(train_data, train_augs)

#load validation dataset
valid_set = load_dataset(valid_data, valid_augs)

#load testing dataset
test_set = load_dataset(test_data, valid_augs)

#check if image is correctly loaded
chosen_idx = 0
img, label = train_set[chosen_idx]
plt.imshow(img.permute(1, 2, 0))
```

# Get Mini-batch of Training and Validation data

```python
trainloader = DataLoader(train_set, batch_size = BATCH_SIZE, shuffle = True)
validloader = DataLoader(valid_set, batch_size = BATCH_SIZE, shuffle = False)
```

# Define Convolution Neural Network

```python
class ConvBlock(nn.Module):
    def __init__(self, in_c, out_c, k_size):
        '''
        Args:
            in_c   : input channel size of Conv2d
            out_c  : output channel size of in Conv2d
            k_size : kernel size of Conv2d
        '''
    
        super(ConvBlock, self).__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_channels = in_c, out_channels = out_c, kernel_size = k_size), 
            nn.ReLU(),  
            nn.MaxPool2d(kernel_size = (4, 4), stride = 2)
        )
        
    def forward(self, x):
        return self.block(x)

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()

        self.feature_extractor = nn.Sequential(
            ConvBlock( 3, 16, (3, 3)),
            ConvBlock(16, 32, (3, 3)),
            ConvBlock(32, 64, (3, 3)),
            nn.Conv2d(in_channels = 64, out_channels = 64, kernel_size = (3, 3)),
            nn.ReLU(),                                                           
        )
        
        self.maxpool = nn.MaxPool2d(kernel_size = (4, 4), stride = 2) 
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(6400, 2048),
            nn.ReLU(),
            nn.Linear(2048, 3)
        )
    
        self.gradient = None
    
    def activations_hook(self, grad):
        self.gradient = grad
    
    def forward(self, images):
        x = self.feature_extractor(images)
        
        h = x.register_hook(self.activations_hook)
        x = self.maxpool(x)
        x = self.classifier(x)
        
        return x
    
    def get_activation_gradients(self):
        return self.gradient
    
    def get_activations(self, x):
        return self.feature_extractor(x)
    
model = Net()
summary(model, (3, 227, 227))
model.to(DEVICE)
```

# Start Training

```python
#define training loop
def train_fcn(dataloader, model, criterion, opt):
    '''
    Args:
        dataloader: mini-batch of training data
        model     : convolutional neural network
        criterion : loss function for optimization
        opt       : optimizer for updating parameters of neural network
    Outputs:
        avg_loss  : average loss of one training epoch
    '''
    
    
    model.train()
    total_loss = 0.0
    n = 0
    for (images, labels) in dataloader:
        images = images.to(DEVICE)
        labels = labels.to(DEVICE)
        
        opt.zero_grad()
        preds = model(images)
        loss = criterion(preds, labels)
        
        loss.backward()
        opt.step()
        
        total_loss += loss.item()
        n += 1
        
    avg_loss = total_loss/n
    
    return avg_loss

#define evaluation loop
def eval_fcn(dataloader, model, criterion):
    '''
    Args:
        dataloader: mini-batch of validation data
        model     : convolutional neural network
        criterion : loss function for optimization
    Outputs:
        avg_loss  : average loss of one training epoch
    '''
    
    model.eval()
    total_loss = 0.0
    n = 0
    for (images, labels) in dataloader:
        images = images.to(DEVICE)
        labels = labels.to(DEVICE)
        
        preds = model(images)
        loss = criterion(preds, labels)
        
        total_loss += loss.item()
        n += 1
        
    avg_loss = total_loss/n
    
    return avg_loss
```

```python
#define the optimizer
opt = torch.optim.Adam(model.parameters(), lr = LR)
#define the loss
criterion = torch.nn.CrossEntropyLoss()

#start training 
best_valid_loss = np.Inf
train_loss_list = [None]*EPOCH
valid_loss_list = [None]*EPOCH

for i in range(EPOCH):
    train_loss_list[i] = train_fcn(trainloader, model, criterion, opt)
    valid_loss_list[i] = eval_fcn(validloader, model, criterion)
    
    #save better model with lower validation loss
    if valid_loss_list[i] < best_valid_loss:
        best_valid_loss =  valid_loss_list[i]
        torch.save(model.state_dict(), 'best_weights.pt')
        print('Found and saved model!')
        
    print(f'epoch {i + 1} training loss: {train_loss_list[i] :.2f}, validation loss : {valid_loss_list[i] :.2f}')
```

# Define GradCam Function 

```python
# define gramcam function

def compute_gradcam(model, image, label_index, size):
    image = image.unsqueeze(0).to(DEVICE)
    preds = model(image)
    _, max_pred = torch.max(preds.cpu(), dim = 1)
    print(f'prediction: {class_dict[max_pred.item()]} \nground truth: {class_dict[label_index]}')
    preds = preds[0][label_index]
    
    preds.backward()
    grads = model.get_activation_gradients()
    weights = torch.mean(grads, dim = (0, 2, 3))
    activations = model.get_activations(image).detach() 
    
    for i in range(activations.shape[1]):
        activations[0, i, :, :] *= weights[i]
    
    z = torch.sum(activations, dim = (0, 1))
    heatmap = nn.ReLU()(z)
    heatmap = (heatmap/torch.max(heatmap)).cpu()
    heatmap = cv2.resize(heatmap.numpy(), (size, size))
    
    return heatmap

```

# Heatmap Visualization 

```python
#load test dataset
image, label = test_set[0]

#generate heatmap of activation based on gradcam technique
heatmap = compute_gradcam(model, image, label, image.shape[1])

#convert to original image
original_image = image.permute(1, 2, 0)*np.array((0.229, 0.224, 0.225)) + np.array((0.485, 0.456, 0.406))

#plot original image and the heatmap
fig, ax = plt.subplots(1, 2)
ax[0].imshow(original_image)
ax[1].imshow(original_image)
ax[1].imshow(heatmap, cmap = 'jet', alpha = 0.7)
```
